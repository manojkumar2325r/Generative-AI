```python
from langchain_community.document_loaders import (
    PyPDFDirectoryLoader,
    WebBaseLoader,
    NotionDirectoryLoader,
)
from langchain_google_community import GoogleDriveLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter   # ✅ fixed
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
print("📄 Loading PDFs...")
pdf_loader = PyPDFDirectoryLoader("Cricket Rules.pdf")
pdf_docs = pdf_loader.load()
```

```python
print("🌐 Loading cricket websites...")
urls = [
    "https://www.icc-cricket.com/about/cricket/history-of-cricket",
    "https://www.espncricinfo.com/story/cricket-formats-explained-test-odi-t20-1177886",
    "https://www.lords.org/mcc/the-laws-of-cricket",
]
try:
    web_loader = WebBaseLoader(urls)
    web_docs = web_loader.load()
    print(f"   ✅ {len(web_docs)} website pages loaded")
except Exception as e:
    print(f"   ⚠️  Website loading failed: {e}")
    web_docs = []
```

```python
all_docs = pdf_docs + web_docs
print(f"\n📦 Total documents loaded: {len(all_docs)}")
```

```python
print("✂️  Chunking documents...")
splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100,
    separators=["\n\n", "\n", ".", " "],
)
chunks = splitter.split_documents(all_docs)
print(f"   ✅ {len(chunks)} chunks created")
```
